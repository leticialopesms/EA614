```python
import numpy as np
import matplotlib.pyplot as plt
from tk import *
from matplotlib.widgets import Slider
```

```python
fig, ax = plt.subplots(2,1)

t = np.arange(0.0, 10, 1e-2)
n = np.arange(0, 11, 1)
A = 1
w_0 = 0
y1 = A*np.cos(w_0*t)
y2 = A*np.cos(w_0*n)
l1, = ax[0].plot(t, y1)
l2 =  ax[1].plot(n, y2, 'ro')

ax[0].title.set_text('Caso Contínuo')
ax[1].title.set_text('Caso Discreto')

ax[0].grid()
ax[1].grid()

ax[0].set_ylim([-1.1,1.1])
ax[1].set_ylim([-1.1,1.1])

ax[1].set_xticks(n)

ax_freq = plt.axes([0.25, .03, 0.50, 0.02])

# Slider
s_freq = Slider(ax_freq, 'Frequência Angular [rad/s]', 0, 4*np.pi, valinit=w_0)

def update(val):
    # amp is the current value of the slider
    w = s_freq.val
    # update curve
    l1.set_ydata(A*np.cos(w*t))
    l2[0].set_ydata(A*np.cos(w*n))
    # redraw canvas while idle
    fig.canvas.draw_idle()

# call update function on slider value change
s_freq.on_changed(update)

plt.show()
```

# Periodicidade


```python
w0 = np.sqrt(2)
n = np.arange(0, 21, 1)
y = np.cos(w0*n)

plt.figure()
plt.stem(n, y)
plt.xlabel('n')
plt.grid()
plt.xticks(n)
plt.ylabel('cos[w0.n]')
```
